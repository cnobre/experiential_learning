```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/vlat_pilot_5.csv')

# Display the first few rows of the DataFrame to understand its structure
df.head()
```

```python
import re

# Provided base column names

base_columns = [
    'ACQ1', 'ACQ2', 'ACQ3', 'ACQ4', 'BCQ1', 'BCQ2','BCQ3', 'BCQ4', 'BuCQ1','BuCQ2', 'BuCQ3', 'BuCQ4', 'BuCQ5', 'BuCQ6', 'BuCQ7',
    'CCQ1', 'CCQ2', 'CCQ3', 'HCQ1', 'HCQ2', 'HCQ3', 'LCQ1', 'LCQ2', 'LCQ3', 'LCQ4','LCQ5', 'PCQ1', 'PCQ2', 'PCQ3',
    'SAQ1', 'SAQ2', 'SAQ3', 'SAQ4', 'SAQ5','SAQ6', 'SAQ7', 'SB100Q1', 'SB100Q2', 'SB100Q3', 'SB100Q4',
    'SBQ1', 'SBQ2', 'SBQ3', 'SBQ4', 'SBQ5', 'SPQ1', 'SPQ2','SPQ3', 'SPQ4', 'SPQ5', 'SQP6', 'TMQ1','TMQ2','TMQ3','AC2Q1','AC2Q2','AC2Q3','AC2Q4', 'AC2Q5']
print(len(base_columns))

# Create a regex pattern to match base column names followed by optional underscore and numbers
pattern = re.compile(r'^(?:' + '|'.join(base_columns) + r')(_\d+)?$')
# pattern_str = r'^(?:' + '|'.join(base_columns) + r')(_\d+)?|^(?:' + '|'.join(base_columns) + r' - Annotate_Id$)'
pattern2 = re.compile(r'^(?:' + '|'.join(base_columns) + r' - annotate_Id$)')
# pattern = re.compile(pattern_str)

# Filter columns based on the pattern
# filtered_df1 = df.loc[:, df.columns[df.columns.str.match(pattern)]]
filtered_df = df.loc[:, df.columns[df.columns.str.match(pattern2)]]
```

```python
# Filter columns based on the pattern
filtered_columns = df.columns[df.columns.str.match(pattern)]
# Store the column names as a list
column_names_list = filtered_columns.tolist()

# Remove the first two rows from the dataframe
df_filtered = filtered_df.iloc[1:]
# df_filtered.to_csv('filtered_file.csv', index=False
df_filtered
```

```python
# Find columns in the dataset that contain the keyword "Annotate"
annotate_related_columns = [col for col in df_filtered.columns if "annotate_Id" in col]

annotate_related_columns
```

```python
# Integrated code to compute correctness rates for each question

# Dictionary to store the correctness rates
integrated_correctness_rates = {}
# print(len(annotate_related_columns))

# Loop through each "Annotate_Id" column
for question_col, annotate_id_col in zip(column_names_list, annotate_related_columns):

    # If the question column exists in the dataset
    if question_col in df_filtered.columns:

        # Count the number of correct answers based on the given criteria
        correct_criteria = (filtered_df[question_col].notna()) & (filtered_df[annotate_id_col].isna())

        # Count the number of correct answers based on the given criteria
        correct_count = correct_criteria.sum()

        # Total answers for the question
        total_answers = filtered_df[question_col].notna().sum()

        # Correctness rate
        # correctness_rate = correct_count / total_answers if total_answers != 0 else 0
        # integrated_correctness_rates[question_col] = correctness_rate
        correctness_rate = correct_count / total_answers if total_answers != 0 else 0
        integrated_correctness_rates[question_col] = correctness_rate

integrated_correctness_rates
# len(integrated_correctness_rates)
```

```python
# Refactor keys in the dictionary
refactored_dict = {}
for key, value in integrated_correctness_rates.items():
    base_key = key.split('_')[0]
    print(base_key, value)
    if base_key in refactored_dict:
        # refactored_dict[base_key].append(value)
        print(base_key, value)
    else:
        refactored_dict[base_key] = value

refactored_dict
```

```python
question_answer_dict = {
 'SBQ3': 'FALSE',
 'SBQ4': 'FALSE',
 'SAQ2': 'FALSE',
 'SAQ5': 'TRUE',
 'SPQ4': 'FALSE',
 'SPQ5': 'FALSE',
 'HCQ2': 'FALSE',
 'TMQ2': 'FALSE',
 'BuCQ3': 'TRUE',
 'BuCQ4': 'FALSE',
 'BuCQ5': 'FALSE',
 'SB100Q3': 'Impossible to answer with this visualization.',
 'PCQ2': 'Samsung',
 'SBQ2': 'Stockholm'
}
```

```python
# Extract the answers for the relevant Question IDs
relevant_answers = df_filtered[list(question_answer_dict.keys())]

# Check the correctness of answers
correctness = relevant_answers.apply(lambda col: col.astype(str).str.lower() == question_answer_dict[col.name].lower())

# Exclude empty entries and 'nan' from the calculation of correctness rate
correctness_rate = correctness.apply(lambda col: col[(col.notna()) & (relevant_answers[col.name].astype(str).str.lower() != 'nan')].mean())

correctness_rate
```

```python
# Re-calculate the correctness_rate as a Pandas Series
correctness_rate_series = correctness.apply(lambda col: col[(col.notna()) & (relevant_answers[col.name].astype(str).str.lower() != 'nan')].mean())

# Convert the Pandas Series to a dictionary
correctness_rate_dict = correctness_rate_series.to_dict()

# Update the refactored_dict with the new correctness_rate_dict
refactored_dict.update(correctness_rate_dict)

refactored_dict
```

```python
integrated_correctness_rates = refactored_dict
```

```python
## maunally correct the data source
integrated_correctness_rates['SAQ4'] = 0.974
```

```python
import matplotlib.pyplot as plt

# Assuming integrated_correctness_rates has been computed
correctness_rates_series = pd.Series(integrated_correctness_rates)

# Sort values for better visualization
correctness_rates_series = correctness_rates_series.sort_values()

# Plot
plt.figure(figsize=(15,8))
correctness_rates_series.plot(kind='bar', color='skyblue')

plt.title('Correctness Rates for Each Question')
plt.xlabel('Questions')
plt.ylabel('Correctness Rate')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```
